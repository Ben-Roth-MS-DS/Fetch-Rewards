## Single Receipt Deep Dives

```python
#look at single receipt id to get a localized view of what the data does/should look like
receipt_id = receipts.id.values[0]

receipts_dive = receipts.loc[receipts.id == receipt_id, ]
receipts_dive
```

Looking at the output of the above table, there are multiple issues with the data, specifically in the fields that appear to originate from the 'rewardsReceiptItemList' column in the receipts.json file. First and foremost, the purchasedItemCount field in the data says that there are 11 items on the receipt, when in reality it appears there are only two items, with the second item being duplicated 10 times. It appears as though the duplicates come from an instance where the User input the data onto the receipt. Generally, a lot of the data is unpopulated, which could present further issues for any in depth analysis.

```python
#Observing receipts with multiple items that were finished
id_fin = receipts.loc[(receipts.purchasedItemCount > 1) & (receipts.rewardsReceiptStatus	== 'FINISHED'), 'id'].values[0]
receipts_fin = receipts.loc[receipts.id == id_fin, ]
receipts_fin
```

Even looking at a receipt that was successfully processed through the Fetch system, there are issues with how the receipt is processed. The item price for the items is incorrect, as three 24 packs of beer would not be $1 each. There also appear to be duplicate items, or at the very least inconsistencies with how duplicate items are processed. 

The first item, the Sargento cheese, appears twice, with one of the instances having the item price listed as $1. Additionally, one of the observations as quantity purchased as two, and the other as one. It would stand to reason that the cheese should all be on one line with a quantity purchased of three if three units were actually purchased. A similar issue is apparent with the miller light packs. 

Furthermore, one of the items on the receipt-level dataset is totalSpent, corresponding to how much the user spent; for this observation, that amount is $14. Further down the dataset, however, there is a semi-populated targetPrice field that corresponds to a much higher value, $77 for the beer specifically.

Finally, the value of total points earned for the receipt, 389.2, does not appear to correspond to the total amount of the pointsEarnedItemInfo	field, which consists of two blank values for the cheese and three values of 30 for the beer.

## Rewards Group Item Duplicates

```python
#Create high level statistics showcasing some of the issues with the data at a larger scale
df_rewardsCount = receipts.\
                    groupby(['id', 'rewardsGroup']).\
                    size().reset_index(name='counts').\
                    sort_values('counts', ascending=False)

df_rewardsCount.head(20)
    
```

```python
df_rewardsCount.\
    groupby('counts').\
    size().\
    reset_index(name = 'Count of Counts').\
    sort_values('counts', ascending=True)
```

The above two tables show several interesting, potential issues with the data. In the first table, we can see that some of the tables show an absurd number of repeated items, intuitively it is fair to assume a majority of people would not buy 15 or 20 packages of Ben & Jerry's ice cream in a single instance. The second table, however, show that instances of these duplicates are rare; a majority of the data contains only single instances of an item being purchased.

## Item Price Analysis

```python
#Create high level statistics showcasing some of the issues with the data at a larger scale
df_itemPrice = receipts.\
                    groupby(['id', 'itemPrice']).\
                    size().reset_index(name='counts').\
                    sort_values('counts', ascending=False)

df_itemPrice.head(20)
```

```python
df_itemPrice.\
    groupby('counts').\
    size().\
    reset_index(name = 'Count of Counts').\
    sort_values('counts', ascending=True).\
    head(10)
```

The above tables highlights another issue, larger issue with the quality of the data; many items on the same receipt and across many receipts costing exactly $1.00, even as we know most items do not cost $1.00 (see the Miller Light example above). This could have larger implications as to how app's points are calculated and awarded.
